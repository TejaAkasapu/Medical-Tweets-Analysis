```
import re
import string
from bs4 import BeautifulSoup

#for tweet in tweet_data:
def preprocess_tweet(tweet):
    # Clean the training data
    # print(tweet)
    
    #Escaping HTML characters
    tweet = BeautifulSoup(tweet).get_text()
    
    #lowercase
    tweet = tweet.lower()
    
    #remove links
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+)|(http?://[^\s]+))','',tweet)
    
    #Remove usernames
    tweet = re.sub('@[^\s]+','', tweet)
    
    # replace hashtags by just words
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet)
    
    #Removal of Punctuation
    tweet = ' '.join(re.sub("[\.\,\!\?\:\;\-\=]", " ", tweet).split())
    
    #correct all multiple white spaces to a single white space
    tweet = re.sub('[\s]+', ' ', tweet)
   
    #remove space at the beginning and the end
    tweet = tweet.strip()
    # print(tweet)
    
    return tweet 
    
        
#     # remove special characters, numbers, punctuations
#     text = text.replace("[^a-zA-Z#]", " ")
#     # Additional clean up : removing words less than 3 chars, and remove space at the beginning and the end
#     text = re.sub(r'\W*\b\w{1,3}\b', '', text)
```

```
def train(training_data_path,validation_data_path):
    print('Training start')
    hyper_params = {"lr": 0.01,
                    "epoch": 2,
                    "wordNgrams": 20,
                    "dim": 20}     
                            
    # Train the model.
    model = fasttext.train_supervised(input=training_data_path, **hyper_params)
    print("Model trained with the hyperparameter \n {}".format(hyper_params))

    model_acc_validation_set = model.test(validation_data_path)
    
    # DISPLAY ACCURACY OF TRAINED MODEL
    text_line =  "validation:" + str(model_acc_validation_set[1]) + '\n'
    print(text_line)
```

```
training_data_path = os.path.join("SE_train.txt")
validation_data_path = os.path.join("SE_validation.txt")
```
